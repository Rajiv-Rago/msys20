```python
#Problem A: POS Machine

class POSmachine:
    
    def __init__(self, name):
        self.priceList = []
        self.salesInvoice = {"ID": ["Product", "Price", "Quantity", "Total"]}
        self.total = 0
        self.invoicesPrinted = 0
        self.name = name #name of store
        
        self.start()
        
    
    def start(self):
        self.printWelcomeMessage()
        self.printAllOptions()
        inp = int(input())
        while inp != 4:
            if inp == 1:
                self.setPriceList(input("Set the price list: "))
            elif inp == 2:
                self.printPriceList()
            elif inp == 3:
                self.makeNewTransaction()
            else:
                print("That is not a valid option.")
            self.printAllOptions()
            inp = int(input())

                
    def printWelcomeMessage(self):
        print("Welcome to {}!".format(self.name))
            
    def printAllOptions(self):
        print("What do you want to do today? (input the number)")
        print("1 - Set the price list")
        print("2 - Show the price list")
        print("3 - Make a new transaction")
        print("4 - Close the POS machine")
    
    def setPriceList(self, file_path):
        try:
            inputFile = open(file_path, "r")

            products = inputFile.read().split("\n")

            if file_path[-3:] == "csv":
                self.priceList = [product.split(",") for product in products]
            elif file_path[-3:] == "txt":
                self.priceList = [product.split("\t") for product in products]

            inputFile.close()
        except:
            print("Invalid File")
        finally:
            print("")
    
    def printPriceList(self):
        for product in self.priceList:
            output = ""
            
            for attribute in product:
                output += "{}  ".format(attribute)
            print(output)
        print("")
        
    def addProductToTransaction(self, product, quantity):
        tempList = self.priceList[product] + [quantity, quantity * int(self.priceList[product][2])]
            
        if tempList[0] in self.salesInvoice:
            self.salesInvoice[tempList[0]][2] += tempList[3]
            self.salesInvoice[tempList[0]][3] += tempList[4]
        else:
            self.salesInvoice[tempList[0]] = tempList[1:]
    
    def makeNewTransaction(self):
        try:
            print("Input the product and quantities (input blank to complete the transaction)")
            buy = input().replace(" ", "").split(",")

            while len(buy[0]) != 0:
                buy = [int(x) for x in buy]
                self.addProductToTransaction(buy[0], buy[1])
                
                buy = input().replace(" ", "").split(",")
                
            self.createSalesInvoice()
            print("")
        except:
            print("Product does not exist\n")
            self.makeNewTransaction()
    
    def createSalesInvoice(self):
        output = "Sales Invoice\n"
        
        #Source: <https://www.w3schools.com/python/gloss_python_loop_dictionary_items.asp>
        #Date: <November 21, 2021>
        #Name of Author/Programmer: W3Schools
        for ID, properties in self.salesInvoice.items():
            line = str(ID)
            
            if type(properties[3]) != str:
                self.total += properties[3]
            
            for item in properties:
                line += "\t{}".format(item)
            output += line + "\n"
            
        output += "Grand Total {}".format(self.total)
        
        self.invoicesPrinted += 1
        outputFile = open("invoice{}.txt".format(self.invoicesPrinted), "w")
        
        outputFile.write(output)
        
        outputFile.close()
        
        self.total = 0
        self.salesInvoice = {"ID": ["Product", "Price", "Quantity", "Total"]}
        print(output.replace("\t", "  "))
```
