```python
#Problem A: Flights

class Flight:
    def __init__(self):
        self.__fuelCost = 0.0
        self.__fuelPrice = 30000
        
        self.regularPrice = 2000
        self.seatsAvailable = 300
        self.totalSales = 0
        self.seatsSold = [0, 0, 0] #[regSeats, discountSeats, premiumSeats]
        
        self.totalCost = 0
        
    def sellSeats(self, regSeats, discountSeats, premiumSeats):
        seats = [regSeats, discountSeats, premiumSeats]
        totalSeats = sum(seats)
        
        if self.seatsAvailable < totalSeats:
            print("Total cost is 0.0")
            
        else:
            self.seatsAvailable -= totalSeats
            for index in range(len(seats)):
                self.seatsSold[index] += seats[index]
            
            self.totalCost = (seats[0] * self.regularPrice) + (seats[1] * self.regularPrice * 0.9) + (seats[2] * self.regularPrice * 1.1)
            self.totalSales += self.totalCost
            print("Total cost is {}".format(self.totalCost))
            
            
    def flyPlane(self):
        self.__fuelCost += self.__fuelPrice
        self.seatsAvailable = 300
        
        
    def changeSeatPrice(self, newCost):
        self.regularPrice = newCost
        
        
    def changeFlightCost(self, newCost):
        self.__fuelPrice = newCost
        
        
    def printReport(self):
        print("----Salse Report----")
        print("Total sales: {}".format(self.totalSales))
        print("Total regular seats sold: {}".format(self.seatsSold[0]))
        print("Total discount seats sold: {}".format(self.seatsSold[1]))
        print("Total premium seats sold: {}".format(self.seatsSold[2]))
        print("Total flight cost: {}".format(self.__fuelCost))
        print("--------------------")


pal = Flight()

pal.sellSeats(30, 25, 40)
pal.printReport()

pal.flyPlane()

pal.sellSeats(50, 100, 100)
pal.printReport()

pal.sellSeats(60, 70, 15)

pal.flyPlane()

pal.printReport()

pal.changeSeatPrice(3000.00)

pal.sellSeats(45, 70, 20)
pal.printReport()

pal.sellSeats(150, 0, 10)

pal.flyPlane()

pal.changeFlightCost(50000)

pal.sellSeats(180, 65, 0)
pal.printReport()

pal.sellSeats(50, 240, 20)

pal.flyPlane()

pal.printReport()
```

```python
#Problem B: Flights, extended

class CancellableSeatsFlight(Flight):
    def __init__(self):
        Flight.__init__(self)
        self.latestSeats = [0, 0, 0]
        self.refundRate = 0.5
    
    def sellSeats(self, regSeats, discountSeats, premiumSeats):
        Flight.sellSeats(self, regSeats, discountSeats, premiumSeats)
        self.latestSeats = [regSeats, discountSeats, premiumSeats]
        
    def cancelSeats(self):
        for index in range(len(self.latestSeats)):
            self.seatsSold[index] -= self.latestSeats[index]
            
        self.totalSales -= self.totalCost * self.refundRate
        self.seatsAvailable += sum(self.latestSeats)
        self.latestSeats = [0, 0, 0]
        
        print("Total refunded sales is {}".format(self.totalCost * self.refundRate))
        
        
    def changeRefundRate(self, newRate):
        self.refundRate = newRate
        
        
ceb = CancellableSeatsFlight()

ceb.sellSeats(30, 80, 80)
ceb.printReport()

ceb.sellSeats(60, 70, 15)
ceb.printReport()

ceb.sellSeats(20, 30, 40)
ceb.printReport()

ceb.cancelSeats()

ceb.flyPlane()

ceb.printReport()

ceb.changeRefundRate(0.4)

ceb.sellSeats(50, 50, 50)
ceb.printReport()

ceb.sellSeats(40, 30, 20)
ceb.printReport()

ceb.cancelSeats()

ceb.flyPlane()

ceb.printReport()
```
