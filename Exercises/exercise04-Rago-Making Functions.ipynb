```python
def countRange(nums, low, high):
    x = 0
    
    for num in nums:
        if num >= low and num < high:
            x += 1
    return x
```

```python
def integerToOrdinal(x):
    ordinals = ["First", "Second", "Third", "Fourth", "Fifth", " Sixth", "Seventh", "Eighth", "Nineth", "Tenth", "Eleventh", "Twelfth"]
    return ordinals[x - 1]
    
    
```

```python
def daysOfChristmas(day):
    ordinals = ["First", "Second", "Third", "Fourth", "Fifth", " Sixth", "Seventh", "Eighth", "Nineth", "Tenth", "Eleventh", "Twelfth"]
    verses_complete = ["And, a partridge in a pear tree.", "Two turtle doves,", "Three french hens,", "Four calling birds,", "Five gold rings,", "Six geese a-laying,", "Seven swans a-swimming,", "Eight maids a-milking,", "Nine ladies dancing,", "Ten lords a-leaping,", "Eleven pipers piping,", "Twelve drummers drumming,"]
    verses_to_include = ""
    
    x = 0
    while x < day:
        if day == 1:
            verses_to_include = "A partridge in a pear tree.\n"
        else:
            verses_to_include += verses_complete[day - x - 1] + "\n"
        x += 1
    
    return "On the {} day of Christmas \nmy true love sent to me: \n{}".format(ordinals[day-1], verses_to_include)

x = 1
while x <= 12:
    print(daysOfChristmas(x))
    x += 1
```
