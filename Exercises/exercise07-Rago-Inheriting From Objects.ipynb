```python
#Problem A: Checking accounts, improved

class BankAccount:
    def __init__(self):
        self.__balance = 0
        
        
    def deposit(self, amount):
        self.__balance += amount
        
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds")
    
    def getBalance(self):
        return self.__balance
        
class CheckingAccount(BankAccount):
    def __init__(self):
        self.__transactions = 0
        BankAccount.__init__(self)
    
    def deposit(self, amount):
        self.__transactions += 1
        BankAccount.deposit(self, amount)
        self.deductFees()
        
    def withdraw(self, amount):
        self.__transactions += 1
        BankAccount.withdraw(self, amount)
        self.deductFees()
    
    def getBalance(self):
        return BankAccount.getBalance(self)
    
    def deductFees(self):
        if self.__transactions >= 3:
            BankAccount.withdraw(self, 10)
            self.__transactions = 0
        
    

check = CheckingAccount()

check.deposit(100)
print(check.getBalance())

check.deposit(100)
print(check.getBalance())

check.withdraw(100)
print(check.getBalance())

check.withdraw(1000)
print(check.getBalance())

check.deposit(1000)
print(check.getBalance())

check.deposit(1000)
print(check.getBalance())
```

```python
#Problem B: Management

class Employee:
    def __init__(self):
        self.work_hours = 40
        self.salary = 30000
        self.vacation_days = 10
        self.vacation_form = "green"
    
    def getHours(self):
        return self.work_hours
    
    def getSalary(self):
        return self.salary
    
    def getVacationDays(self):
        return self.vacation_days
    
    def getVacationForm(self):
        return self.vacation_form

class Lawyer(Employee):
    def __init__(self):
        Employee.__init__(self)
        self.vacation_days += 5
        self.vacation_form = "blue"
    
    def sue(self):
        return "I'll see you in court!"
    
class Marketer(Employee):
    def __init__(self):
        Employee.__init__(self)
        self.salary += 5000
        
    def advertise(self):
        return "Act now, while supplies last!"
    
class InternationalLawyer(Employee):
    def __init__(self):
        Employee.__init__(self)
        self.salary *= 1.20
        self.salary = int(self.salary)
        self.vacation_days += 2
        self.vacation_form = "blueblueblue"


employee = Employee()
lawyer = Lawyer()
marketer = Marketer()
international_lawyer = InternationalLawyer()

print(employee.getHours())
print(employee.getVacationForm())
print(lawyer.getVacationDays())
print(lawyer.getVacationForm())
print(marketer.getSalary())
print(marketer.advertise())
print(international_lawyer.getSalary())
print(international_lawyer.getVacationForm())
```
